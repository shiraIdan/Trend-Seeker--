```python
import pandas as pd

df = pd.read_csv("database/recipes.csv")
df[['RecipeId', 'Name', 'AuthorName', 'RecipeInstructions', 'Description']].head()
```

```python
# 1. Define the transform_text function
def transform_text(text):
    # Split the text at every comma
    portions = text.split(',')
    
    # Remove leading and trailing whitespace from each split portion
    cleaned_portions = [portion.strip() for portion in portions]
    
    # Join the split portions with newline characters
    result = '\n'.join(cleaned_portions)
    
    return result
```

```python
#  Removing Duplicates
df.drop_duplicates(inplace=True)

#  Handling Missing Values
df['Name'].fillna('Unknown', inplace=True)
df['AuthorName'].fillna('Unknown', inplace=True)
df['RecipeInstructions'].fillna('Unknown', inplace=True)
df['Description'].fillna('Unknown', inplace=True)

#  Applying the transform_text function to the RecipeInstructions column
df['RecipeInstructions'] = df['RecipeInstructions'].apply(transform_text)

#  Normalizing Data: Converting ingredient names to lowercase and remove punctuation
df['RecipeInstructions'] = df['RecipeInstructions'].str.lower()
df['RecipeInstructions'] = df['RecipeInstructions'].str.replace('[^\w\s]', '') 


df[['RecipeId', 'Name', 'AuthorName', 'RecipeInstructions', 'Description']].head()
```
